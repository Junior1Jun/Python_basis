```python
cospi=req.get('https://finance.naver.com/sise/sise_quant.naver')
soup=bs(cospi.text,'html.parser')
tltle=soup.select("a.tltle")
numbers=soup.select("td.number")
num=1
for i in range(len(tltle)):
    print(str(num)+". "+str(tltle[i].text))    
    print(str(numbers[10*i].text+'\n\n'))
    num+=1
```

```python
# 데이터 프레임으로 출력하기
indexes=[]
for i in tltle:
    indexes.append(i.text)
prices=[]
for i in range(100):
    prices.append(numbers[i*10].text)
datas={"종목명":indexes,'현재가':prices}    
import pandas as pd
prices_df=pd.DataFrame(datas)
prices_df.set_index("종목명",inplace=True)
prices_df
```

```python
bugs=req.get('https://music.bugs.co.kr/chart')
soup=bs(bugs.text,'html.parser')

ranking=soup.select('div.ranking>strong')
ranker=[]
for i in ranking:
    ranker.append(i.text)
    
titles=soup.select('p.title>a')
title_list=[]
for i in range(len(ranker)):
    title_list.append(titles[i].text)


arts=soup.select('p.artist>a')
artist_check=soup.select('a.artistTitle')

for i in artist_check:
    arts.remove(i)
arts_list=[]
for i in arts:
    arts_list.append(i.text)

result={"순위":ranker,"곡":title_list,"아티스트":arts_list}
result_df=pd.DataFrame(result)
result_df.set_index("순위",inplace=True)
result_df
```

```python
result_df.to_csv("result.csv",encoding='euc-kr')
```

```python
movie=req.get('https://movie.naver.com/movie/point/af/list.naver')
soup=bs(movie.text,'html.parser')

titles=soup.select("a.movie.color_b")
tit_list=[]
for i in titles:   
    tit_list.append(i.text)
    
scores=soup.select("div.list_netizen_score>em")
sco_list=[]
for i in scores:
    sco_list.append(i.text)


review=soup.select('td.title')
rev_list=[]
for i in range(len(review)):
    rev_list.append(review[i].text.split('\n')[5])

    
result={"영화제목":tit_list,"평가":sco_list,"리뷰":rev_list}
result_df=pd.DataFrame(result)
result_df.set_index("영화제목",inplace=True)
result_df
```

```python
view=soup.select('td.title')
title_list=[]
review_list=[]
for i in view:
    string=i.text.split('\n\n')
    title_list.append(string[0].replace('\n',''))
    review=string[2].replace('\n','')
    review=review.replace('\t','')
    review=review.replace('신고','')
    
    review_list.append(review)
review_list
```

```python
title_list=[]
review_list=[]
rating_list=[]

for i in range(1,6):
    url='https://movie.naver.com/movie/point/af/list.naver?&page={}'.format(i)
    res=req.get(url)
    
    soup=bs(res.text,'html.parser')
    
    #별점
    score=soup.select('td.title>div>em')
    for i in score:
        rating_list.append(i.text)
    
    #제목/리뷰
    view=soup.select('td.title')
    
    for i in view:
        string=i.text.split('\n\n')
        title_list.append(string[0].replace('\n',''))
        review=string[2].replace('\n','')
        review=review.replace('\t','')
        review=review.replace('신고','')
        
        review_list.append(review)
```

```python
len(title_list), len(review_list), len(rating_list)
```

```python
result={"영화제목":title_list,"평가":rating_list,"리뷰":review_list}
result_df=pd.DataFrame(result)
result_df.set_index("영화제목",inplace=True)
result_df
```
