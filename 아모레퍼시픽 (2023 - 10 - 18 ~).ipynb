```python
from selenium import webdriver
from selenium.webdriver import ActionChains

from selenium.webdriver.common.keys import Keys
# 웹 사이트 내에서 마우스 클릭 or 키보드 입력과 같은 이벤트를 발생시키는 모듈

from selenium.webdriver.common.by import By
# 웹 크롤링을 할 때, 크롤링 대상 지정방법을 정해주는 모듈

import time
# 페이지 로딩을 기다리는 데 사용되는 모듈
```

## 1. 홈페이지에서 데이터 크롤링

```python
num = int(input('원하시는 행의 개수 입력 : '))
```

```python
num
```

```python
# 네이버 열기.
driver = webdriver.Chrome()
driver.maximize_window()
act = ActionChains(driver)
URL = 'https://www.naver.com'
driver.get(URL)
time.sleep(1)



# 네이버 검색 창에 '롯데백화점몰'을 입력.
element = driver.find_element(By.CSS_SELECTOR,
                             '#query')
element.send_keys('롯데백화점몰')
element.send_keys(Keys.ENTER)
time.sleep(1)



# '롯데백화점몰'의 홈페이지로 이동.
element = driver.find_element(By.CSS_SELECTOR,
                              '#main_pack > section.sc_new.sp_nsite._project_site_channel_root._section_nsite_._sp_ntotal._prs_vsd_bas > div > div > div.nsite_tit > div > div.nsite_name > a '
                             )
link = element.get_attribute('href')
driver.get(link)
time.sleep(1)
act.send_keys(Keys.ESCAPE).perform()



# 롯데백화점몰 홈페이지의 검색 창에 '아모레퍼시픽' 입력.
element = driver.find_element(By.CSS_SELECTOR,
                              '#headerSearchId')
element.send_keys('아모레퍼시픽')
element.send_keys(Keys.ENTER)
time.sleep(1)



# '리뷰 많은 순' 글자 클릭
element = driver.find_element(By.CSS_SELECTOR,
                              '#s-search-app > div.srchResultProductArea > div.srchResultSortArea > div > div:nth-child(2) > ul > li:nth-child(7) > label > span'
                             )
element.click()
time.sleep(1)



# 빈 리스트 지정
price_group = []
review_group = []
name_group = []


# 40개만 출력해보기
for i in range(num):
    element = driver.find_element(By.XPATH,
                                  '/html/body/div[1]/div/main/div[2]/section/div[2]/section[2]/div[1]/div[2]/div[1]/ul/li[{}]/div/div[2]/a'.format(i+1)
                                 )
    element.send_keys(Keys.ENTER)
    time.sleep(1.5)
    
    driver.switch_to.window(driver.window_handles[-1])
    price = driver.find_element(By.CSS_SELECTOR,
                                  '#stickyTopParent > div.productDetailTop > div:nth-child(2) > div > div.priceInfo > div > span')
    time.sleep(1)
    review = driver.find_element(By.CSS_SELECTOR,
                                 '#stickyTopParent > div.productDetailTop > section > div.pd-widget1__info-box > ul > li > button > span.pd-widget1__product-review-staring'
                                 )
    time.sleep(1)
    name = driver.find_element(By.CSS_SELECTOR,
                               '#stickyTopParent > div.productDetailTop > section > div.pd-widget1__info-box > h2'
                               )
    
    price_group.append(int(price.text.split('\n')[1].replace(',','')))
    review_group.append(float(review.text.split('\n')[1]))
    name_group.append(name.text)
    
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(1)
```

# 2. 가져온 데이터를 데이터프레임화 시키기

```python
import pandas as pd
```

```python
df = pd.DataFrame({"price" : price_group,
                   "review"  : review_group},
                   index = name_group)
df.index.name = 'Product Name'
```
