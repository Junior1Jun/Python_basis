```python
import numpy as np
import matplotlib.pyplot as plt # 그래프와 관련된 라이브러리 생성
%matplotlib inline               
# %matplotlib inline는 해당 브라우저에서 바로 그림을 볼  수 있게 해주는 코드
# inline은 브라우저 내부를 의미함.
```

```python
x =np.arange(8)
y=[5,2,3,6,8,7,5,7]
plt.plot(x,y,ls="-.",lw=5,marker='*',color='y',mec='c',mfc='w',ms=20)
plt.show()
```

```python
x =np.arange(8)
y=[5,2,3,6,8,7,5,7]
plt.plot(x,y,ls="-.",lw=2,marker='<',color='m',mec='k',mfc='k',ms=10)
plt.show()
```

```python
x =np.arange(8)
y=[5,2,3,6,8,7,5,7]
plt.plot(x,y,ls="-.",lw=2,marker='<',color='m',mec='k',mfc='k',ms=10)
plt.xlim(-2,10)  # x축의 범위를 지정해줌.
plt.ylim(-5,12)  # y축의 범위를 지정해줌.
plt.show()
```

```python
x =np.arange(8)
y=[5,2,3,6,8,7,5,7]
plt.plot(x,y,ls="-.",lw=2,marker='<',color='m',mec='k',mfc='k',ms=10)
plt.xlim(-2,10)  # x축의 범위를 지정해줌.
plt.ylim(-5,12)  # y축의 범위를 지정해줌.
plt.grid()       # 그리드 그리기 (그래프(선) 뒤에 가로세로줄을 의미함.)
plt.show()
```

```python
plt.plot(x,y,ls="-.",lw=5,marker='o',mec='g',mfc='r',mew=5,ms=15)
plt.xticks([0,3,6])  # x축의 눈금을 지정
plt.yticks([1,5,9])  # y축의 눈금을 지정
plt.show()
```

```python
# 여러 개 plot 그리기

x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]
plt.plot(x,label='가') # 범례 주석을 달기 위해 plot에 label을 지정해주어야 한다.
plt.plot(y,label='나')
plt.plot(z,label='다')
plt.legend(loc=4)      # 지정된 label들을 바탕으로 범례 주석을 만들어 띄운다.
plt.title("범례 표시") # 그래프의 제목을 지정해준다.
plt.xlabel('x축')
plt.ylabel('y축').set_rotation(0) # 회전율을(각도) 0으로 만듬
plt.show()
```
